# DALLAS & AUSTIN ALL PETS 

```python
austin_pets = pd.read_csv('../data/merge_pets_total_austin.csv')
austin_pets = austin_pets.sort_values('Adopted Total Count', ascending = False)
austin_pets
```

```python
dallas_pets = pd.read_csv('../data/merge_pets_total_dallas.csv')
dallas_pets
```

```python
dallas_austin_pets = pd.merge(austin_pets, dallas_pets, on = ('Animal Type'), how= 'outer')
dallas_austin_pets['Total Euthanized'] = dallas_austin_pets['Euthanized Total Count_x'] + dallas_austin_pets['Euthanized Total Count_y']
dallas_austin_pets['Total Adopted'] = dallas_austin_pets['Adopted Total Count_x'] + dallas_austin_pets['Adopted Total Count_y']
dallas_austin_pets = dallas_austin_pets[['Animal Type', 'Total Euthanized', 'Total Adopted']]
dallas_austin_pets
```

```python
dallas_austin_pets.to_csv('../data/dallas_austin_pets.csv', index=False)
```

# Dallas & Austin Dogs Rates

```python
austin_year = pd.read_csv('../data/merge_pets_year_austin.csv')
austin_year
```

```python
dallas_year = pd.read_csv('../data/merge_pets_year_dallas.csv')
dallas_year
```

```python
dallas_austin_year = pd.merge(austin_year, dallas_year, on = ('Animal Type', 'Year'), how= 'outer')
dallas_austin_year['Total Dog Euthanized'] = dallas_austin_year['Euthanized Animal Count_x'] + dallas_austin_year['Euthanized Animal Count_y']
dallas_austin_year['Total Dog Adopted'] = dallas_austin_year['Adopted Animal Count_x'] + dallas_austin_year['Adopted Animal Count_y']
dallas_austin_year = dallas_austin_year[['Year', 'Animal Type', 'Total Dog Euthanized', 'Total Dog Adopted']]
dallas_austin_year
```

```python
dallas_austin_year.to_csv('../data/dallas_austin_year.csv', index=False)
```

# Dallas & Austin dog breeds

```python
dallas_dogs = pd.read_csv('../data/merge_dogs_dallas.csv')
dallas_dogs.head(12)
```

```python
austin_dogs = pd.read_csv('../data/merge_dogs_austin.csv')
austin_dogs.head(12)
```

```python
dallas_austin_breed = pd.merge(dallas_dogs, austin_dogs, on='Breeds', how= 'outer')
dallas_austin_breed['Total Breed Euthanized'] = dallas_austin_breed['Euthanized Breed Count_x'] + dallas_austin_breed['Euthanized Breed Count_y']
dallas_austin_breed['Total Breed Adopted'] = dallas_austin_breed['Adopted Breed Count_x'] + dallas_austin_breed['Adopted Breed Count_y']
dallas_austin_breed = dallas_austin_breed[['Breeds', 'Total Breed Euthanized', 'Total Breed Adopted']]
dallas_austin_breed = dallas_austin_breed.dropna()
dallas_austin_breed
```

```python
dallas_austin_breed.to_csv('../data/dallas_austin_breed.csv', index=False)
```
